```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from imblearn.combine import SMOTEENN
```

```python
df = pd.read_csv('tel_churn.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0',axis =1)
```

```python
df.head()
```

```python
#Creating X and Y Variable

X  =df.drop('Churn', axis = 1)
X
```

```python
y= df['Churn']
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)     
```

### Decision Tree Classifier


```python
model_dt = DecisionTreeClassifier(criterion='gini', random_state = 100, max_depth =6, min_samples_leaf = 8)
```

```python
model_dt.fit(X_train,y_train)
```

```python
y_pred = model_dt.predict(X_test)
```

```python
y_pred
```

```python
model_dt.score(X_test,y_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test , y_pred, labels =[0,1]))
```

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(X,y)
```

```python
xr_train,xr_test,yr_train,yr_test = train_test_split(X_resampled, y_resampled, test_size = 0.2)
```

```python
model_dt_resampled = DecisionTreeClassifier(criterion = 'gini', random_state = 100, max_depth  =6, min_samples_leaf = 8)
```

```python
model_dt_resampled.fit(xr_train, yr_train)
```

```python
y_pred_smot = model_dt_resampled.predict(xr_test)
```

```python
print(classification_report(yr_test, y_pred_smot, labels=[0,1]))
```

### Random Forest Classifier

```python
model_rc = RandomForestClassifier(n_estimators = 100, criterion ='gini', random_state = 100, min_samples_leaf = 8)
model_rc.fit(X_train,y_train)
y_pred_rc = model_rc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_rc, labels = [0,1]))
```
